#  Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Sorting out required columns for Analysis

```python
df=df[['eventid', 'iyear', 'imonth', 'country', 'region','provstate','city','crit1', 'crit2', 'crit3','success', 'suicide', 'attacktype1','targtype1','natlty1','gname','guncertain1','claimed','weaptype1','nkill','nwound']]
df.head()
```

```python
df.isnull().sum()
```

# Combining Two Columns i.e., 'nkill' and 'nwound' into a new column 'casualities'

```python
df['nkill']=df['nkill'].fillna(0)
df['nwound']=df['nwound'].fillna(0)
df['casualities']=df['nkill']+df['nwound']
df.isnull().sum()
```

```python
print(f"""There are {df.country.nunique()} Countries from {df.region.nunique()} Regions covered in the dataset terrorist attack data in {df.claimed.nunique()} years from {df.iyear.min()} to {df.iyear.max()}. Overall {df.index.nunique()} terrorist attacks are recorded here which caused about {df.casualities.sum()} Casualities Consisted of {df.nkill.sum()} Kills and {df.nwound.sum()} Wounded.
""")
```

# Data Visualization

```python
yearc=df[['iyear','casualities']].groupby('iyear').sum()
yearc.plot(kind='bar',color='red',figsize=(15,6))
plt.title("Casualities")
plt.xlabel('Years',fontsize=15)
plt.ylabel('No. of Casualities',fontsize=15)
plt.show()
```

```python
df['attacktype1'].value_counts().plot(kind='bar',figsize=(20,10),color='blue')
plt.xticks(rotation=50)
plt.xlabel('Attack Type',fontsize=20)
plt.ylabel('Number of attacks')
plt.title('Number of attacks')
plt.show()
```

```python
plt.subplots(figsize=(20,10))
sns.countplot(df['targtype1'],order=df['targtype1'].value_counts().index,palette='gist_heat',edgecolor=sns.color_palette("crest"));
plt.xticks(rotation=90)
plt.xlabel('Attack type',fontsize=20)
plt.ylabel('count')
plt.title('Type of attack')
plt.show()
```

```python
plt.subplots(figsize=(15,6))
country_attacks = df.gname.value_counts()[:15].reset_index()
country_attacks.columns = ['gname', 'Total Attacks']
sns.barplot(x = country_attacks.gname, y = country_attacks['Total Attacks'], palette='OrRd_r', edgecolor=sns.color_palette('dark', 10))
plt.xticks(rotation = 90)
plt.title("Number of Total Attacks in Each Country")
plt.show()
```
